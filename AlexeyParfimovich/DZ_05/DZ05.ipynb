#### Определение класса Alphabet
1. Создать класс Alphabet
2. Создать метод \_\_init\_\_(), внутри которого будут определены два динамических свойства: 1) lang - язык и 2) letters - список букв. Начальные значения свойств берутся из входных параметров метода.
3. Создать метод print(), который выведет в консоль буквы алфавита
4. Создать метод letters_num(), который вернет количество букв в алфавите

```python
class Alphabet():
    def __init__(self, lang:str, letters:set):
        self.lang: lang
        self.letters = set(letters)

    def __str__(self):
        return ",".join(sorted(self.letters))

    def print(self):
        print(f'{self.lang}: {self.__str__()}')

    def letters_num(self):
        return len(self.letters)
```

#### Определение класса AlphabetEng

7. Добавить приватное статическое свойство __letters_num, которое будет хранить количество букв в алфавите.

9. Переопределить метод letters_num() - пусть в текущем классе классе он будет возвращать значение свойства \_\_letters_num.
10. Создать статический метод example(), который будет возвращать пример текста на английском языке.

```python
import string

class AlphabetEng(Alphabet):

    @staticmethod    
    def text_example():
        return "Pack my box with five dozen liquor jugs".upper()

    def __init__(self):
        super().__init__('EN', string.ascii_uppercase)
        self.__letters_num = super().letters_num()

    def letters_num(self):
        #return super().letters_num()
        return self.__letters_num

    def is_en_letters(self, letters:str):
        for i in letters:
            if i.upper() not in self.letters:
                 return False
        return True 
```
