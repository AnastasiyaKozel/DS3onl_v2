## Домашнее задание №7

Написать декоратор, оптимизирующий работу декорируемой функции.  
Декоратор должен сохранять результат работы функции на ближайшие три запуска и вместо выполнения функции возвращать сохранённый результат.  
После трёх запусков функция должна вызываться вновь, а результат работы функции — вновь кешироваться.  

```python
# Декоратор
def buff_func_output(buff_len = 3):
    '''
    Декоратор кэширует вывод функции в очередь(list).
    Аргументы:
    buff_len -- Глубина очереди кеширования (default 3)
    '''
    def func_launcher(func):
        def wrapper(*args):
            buffer.append(func(*args))
            if len(buffer) > buff_len:
                return buffer.pop(0)
            else:
                return f"the function output been buffered, the length of the buffer = {len(buffer)}"
        return wrapper
    buffer = []
    return func_launcher


# Тестовая функция
@buff_func_output(3)
def my_func(*args):
    '''
    Функция возвращает перечень полученных аргументов в виде строки с разделителем ";"
    Аргументы:
    *args - произвольный перечень значений любого типа
    '''
    return ';'.join(str(a) for a in args) 


# Тестирование функции:

print(my_func(1,2,3))

print(my_func(2,3,4))

print(my_func(3,4,5))

print(my_func(5,6,7))

print(my_func(6,7,8))

print(my_func(7,8,9))
```
