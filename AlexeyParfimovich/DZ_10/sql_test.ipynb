```python
import pandas as pd
import numpy as np
import sqlite3

t = pd.DataFrame({'purpose':['машина','машина','машина','на машину','на покупку машины',
                             'автомобиль','на возвращение 2007', 
                             'на свадьбу','свадьба','свадьба','свадьба','для свадьбы',
                             'недвижимость','на покупку недвижимости']})

con = sqlite3.connect('db') #соединяемся с виртуальной базой db

t.to_sql('purposes',con,index=False,if_exists='replace')

# Напишите запрос который бы строил новый столбец и менял бы в нем слова которые содержат %свадьб% на свадьба
# '%авто%' на машина
# '%недвиж%' на недвижимость
# если нет совподений то 'другое'


sql = '''select purpose,
            case 
                when purpose like "%свадьб%" then 'свадьба'
                when purpose like "%авто%" then 'машина'
                when purpose like "%недвиж%" then 'недвижимость'
            else 'другое' end as category
        from purposes'''
pd.read_sql(sql,con)
```
