```python
import matplotlib.pyplot as plt
import numpy as np
```

```python
y_value=[] #пустой список для значений у
x = -5 
while -5 <= x <= 5:
    y = (-26 * (x ** 2)) + (25 * x) - 9
    x += 0.1
    y_value.append(y) #запись всех значений у в список

print(max(y_value), min(y_value))


```

```python
fig, ax = plt.subplots()

y_1=[]
x_1=[]
x = -5

while -5 <= x <= 5:
    y = (-26 * (x ** 2)) + (25 * x) - 9
    x += 0.1
    x_1.append(x)
    y_1.append(y)

ax.set_title('y = -26*x**2+25*x-9')
ax.plot(x_1, y_1, color = 'red')
ax.grid()
ax.set_xticks(np.arange(-6, 7, 1))
ax.set_yticks(np.arange(-900, 10, 100))

plt.show()
```
