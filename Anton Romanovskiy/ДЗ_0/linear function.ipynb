```python
import matplotlib.pyplot as plt
```

```python
fig, ax = plt.subplots()

ax.set_title('Addiction') 
ax.set_xlabel('Celsius')
ax.set_ylabel('Fahrenheit')

ax.grid()  


y = []
x = [0, 3, 8]
for i in x:
    b = (9/5 * i) + 32
    y.append(b)
ax.plot(x, y, color = 'red', marker = 'o')

plt.show()
```
