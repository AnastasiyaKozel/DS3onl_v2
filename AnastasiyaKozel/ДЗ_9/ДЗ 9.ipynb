2. Библиотеки: OpenCV, pillow, scikit-image, tf.image. 
3. Аугментация (вращение, растяжение/сжатие, увеличение/уменьшение). Количество изображений должно увеличиться минимум в 3 раза.


```python
#устанавливаем требуемые библиотеки
import cv2
import os
import glob
from PIL import  Image,ImageFilter
```

```python
# Я так и не дошла до сохранения на гугл диск. Сделала так
images = [Image.open(file) for file in glob.glob(r"C:\Users\37533\Desktop\jupyter\car\*.jpeg")]
```

```python
# уменьшение разрешения картинок и сохранение их в отдельный список и папку
resized_imgs = []
directory = r"C:\Users\37533\Desktop\jupyter\car\New"
filename = 0
size = 256, 256

os.chdir(directory)

for img in images:
    img.thumbnail(size)
    resized_imgs.append(img)
    filename += 1
    img.save(str(filename) + '.jpg')
img
```

### Метод rotate() библиотеки Python Pillow
Принимает число градусов в качестве параметра и поворачивает изображение в направлении против часовой стрелки

```python
filename = 0

for img in resized_imgs:
    im2 = img.rotate(45)
    filename += 1
    img.save(str(filename) + '_rotate1' + '.jpg')
im2
```

```python
filename = 0

for img in resized_imgs:
    im2 = img.rotate(180)
    filename += 1
    img.save(str(filename) + '_rotate2' + '.jpg')
im2
```

#### BLUR 
Размытое изображение

```python
filename = 0
for img in resized_imgs:
    im2=img.filter(ImageFilter.BLUR)
    filename += 1
    img.save(str(filename) + '_BLUR' + '.jpg')
im2
```

#### CONTOUR
Контурная фильтрация, извлечение информации о контуре в изображении

```python
filename = 0
for img in resized_imgs:
    im2=img.filter(ImageFilter.CONTOUR)
    filename += 1
    img.save(str(filename) + '_CONTOUR' + '.jpg')
im2
```

#### DETALL
Фильтрация улучшения деталей, он покажет детали на картинке.

```python
filename = 0
for img in resized_imgs:
    im2=img.filter(ImageFilter.DETAIL)
    filename += 1
    img.save(str(filename) + '_DETAIL' + '.jpg')
im2
```

#### EDGE_ENHANCE
Улучшение краев. Фильтрация по краям - метод улучшения изображения, который выделяет, усиливает и улучшает границы и контуры между различными серыми областями изображения. После обработки граница и край появляются на изображении как внезапное изменение уровня серого изображения, которое используется для улучшения способности распознавания человеческого глаза.

```python
filename = 0
for img in resized_imgs:
    im2=img.filter(ImageFilter.EDGE_ENHANCE)
    filename += 1
    img.save(str(filename) + '_EDGE_ENHANCE' + '.jpg')
im2
```

#### FIND_EDGES
Поиск по краю. Фильтрация для поиска информации о краях позволит найти информацию о краях на изображении.

```python
filename = 0
for img in resized_imgs:
    im2=img.filter(ImageFilter.FIND_EDGES)
    filename += 1
    img.save(str(filename) + '_FIND_EDGES' + '.jpg')
im2
```

```python
filename = 0
for img in resized_imgs:
    im2 = img.convert('L')
    filename += 1
    img.save(str(filename) + '_FIND_EDGES' + '.jpg')
im2
```
