### Создаем список файлов из изображений

```python
#импортируем требуемый библиотеки
import requests
from bs4 import BeautifulSoup
import time
import requests
import concurrent.futures
from threading import Thread
from multiprocessing import Pool, cpu_count


#Сохраним URL в переменную
cat_100 = 'https://bipbap.ru/jivotnie/samye-krasivye-foto-kotikov-100-foto.html'
#Отправим GET()-запрос на сайт и сохраним полученное в переменную 'page'
page = requests.get(cat_100)
#Проверим подключение
print(page.status_code)
```

```python
#создадим пустой лист
list_cat = [] 
soup = BeautifulSoup(page.text, "html.parser")
#Если вызвать print(soup) вылезет весь html-код нашей страницы.

#Воспользуемся функцией поиска в BeautifulSoup4
for img in soup.findAll('img'):
    list_cat.append(img.get('src'))
#в list_cat есть лишние ссылки
#необходимо отфильтровать список от лишнего
list_cat_picture=[]
for img1 in list_cat:
    if 'https:' in img1:
        list_cat_picture.append(img1)
#смотрим сколько картинок        
len(list_cat_picture)  
```

### Последовательный метод

```python
def request_cute_cats(cat_picture_url):
    response = requests.get(url=cat_picture_url)
    return cat_picture_url 

start_1 = time.time()
for url in list_cat_picture:
    print(request_cute_cats(cat_picture_url=url))
print("Последовательный метод:", time.time() - start_1)
```

### ThreadPoolExecutor

```python
start_2 = time.time()
with concurrent.futures.ThreadPoolExecutor(28) as executor:
    futures = []

    for url in list_cat_picture:
        futures.append(executor.submit(request_cute_cats, cat_picture_url=url))
    
    for future in concurrent.futures.as_completed(futures):
        print(future.result())

print("Threaded time:", time.time() - start_2)
```

### Thread

```python
def request_images(from_ind, count):
    for i in range(from_ind, from_ind + count):
        print(list_cat_picture[i])
        
start_3 = time.time()          
thread_count = 12
count = len(list_cat_picture) // thread_count


threads = [Thread(target= request_images , args=(i * count, count)) for i in range(thread_count)]


for thread in threads:
    thread.start()
    thread.join()

    
        
print("Threaded time:", time.time() - start_3)    
```

### Multiprocessing 

```python
from multiprocessing import Pool

def f(cat_picture):
    for cat in cat_picture:
        print(cat)
start_4 = time.time()
if __name__ == '__main__':
    with Pool(3) as p:
        print(p,f(list_cat_picture))
print("Multi time:", time.time() - start_4)  
```
